```python
from deepface import DeepFace
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
 img = cv2.imread("FER//img3.jpg")
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB) )
```

```python
pred = DeepFace.analyze(img)
```

```python
pred
```

```python
pred['dominant_emotion']
```

## We are trying to draw rectangule across the face

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

faces = faceCascade.detectMultiScale(gray1,1.1,4)

# Draw a Rect aronud the Faces
for(x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,225,0),2)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB) )
```

## Overlay text on it

```python
font = cv2.FONT_HERSHEY_SIMPLEX
# use putText() for  test

cv2.putText(img,
           pred['dominant_emotion'],
           (0,50), font, 1, (0,0,255), 2 , cv2.LINE_4);
```
